```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
sns.set_style('darkgrid')
sns.set_context('poster', font_scale=1)

from time import sleep
from datetime import timedelta, datetime, date
import requests
```

```python
me = pd.read_csv('../data/portfolio.csv')
me['total_BTC'] = me.units * me.price_BTC
```

```python
me
```

```python
price_btc = []
price_usd = []
rank = []

for i in me.asset:
    url = 'https://api.coinmarketcap.com/v1/ticker/' + i.lower().replace(' ', '-')
    response = requests.get(url)
    print (i, response.status_code)
    j = response.json()[0]
    sleep(1)
    
    price_btc.append(j['price_btc'])
    price_usd.append(j['price_usd'])
    rank.append(j['rank'])
```

```python
me['current_price_btc'] = price_btc
me['current_price_usd'] = price_usd

me.current_price_btc = me.current_price_btc.astype('float')
me.current_price_usd = me.current_price_usd.astype('float')

me['current_usd_total'] = me.units * me.current_price_usd
me['change'] = (me.current_price_btc - me.price_BTC) / me.price_BTC * 100

me['rank'] = rank
me.set_index('rank', inplace=True)
me.reset_index(inplace=True)
```
